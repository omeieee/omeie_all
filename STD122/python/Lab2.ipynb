```python
math_data = {"multiplier": 2}
print("--- The Doubling Machine ---")
number = float(input("Enter any number to double: "))
print("Enter any number to double:",number)
twice_me = lambda x: x * math_data["multiplier"]
print(f"Result: {twice_me(number)}")
```

```python
def frequecy(x):
    freg_dict = {}
    for num in x:
        num_int = int(num)
        if num_int in freg_dict:
            freg_dict[num_int] += 1
        else:
            freg_dict[num_int] = 1
    return freg_dict

number = input("Enter number separated by spaces: ")
num_split = number.split()
print("Enter number separated by spaces:",number)
result = frequecy(num_split)
print(f"Frequency Dictionary: {result}")
```

```python
def c_gravitational_force(planet, mass):
    gravity_dict = {
        "Earth": 9.8,
        "Moon": 1.6,
        "Mars": 3.7,
        "Saturn": 9.0,
        "Jupiter": 23.1}
    if planet in gravity_dict:
        g_value = gravity_dict[planet]
        force = mass * g_value
        return force
    else:
        return None

planet_input = input("Enter the planet (Earth, Moon, Mars, Saturn, Jupiter): ")
print("Enter the planet (Earth, Moon, Mars, Saturn, Jupiter):",planet_input)
mass_input = float(input("Enter the object's mass in kg: "))
print("Enter the object's mass in kg:",mass_input)
force_result = c_gravitational_force(planet_input, mass_input)
if force_result is not None:
    print(f"Gravitational Force on {planet_input}: {force_result:.2f} N")
else:
    print("Error: Planet not found in the dictionary.")
```
