**Variable Length Arguments**

````python
def test(*args):
  n = args[0]
  for i in args:
    if i < n:
      n = i
  return n
test(4,1,2,3,4)
````

## **Recursion**
A function calling itself is known as recursion.

````python
def rec(n):
  if n == 1:
    return n
  else:
    return n*rec(n-1)
n = 5
if n < 0:
  print('no factorial for negative numbers')
elif n == 0:
  print('The factorial of 0 is 1')
else:
  print('The factorial of', n, 'is', rec(n))
````

````python
def fib(n):
  if n <= 1:
    return n
  else:
    return(fib(n-1) + fib(n-2))
n = 8
if n <= 0:
  print('enter a positive integer:')
else:
  print('Fibonacci sequence:')
  for i in range(n):
    print(fib(i))
````

**Problem 2** 

Write:

A lambda is_vowel(ch) that returns True if ch is a vowel (a, e, i, o, u) case-insensitive.

A normal function count_vowels(s) that counts vowels in a string without using list/tuple/dict.

````python
#Problem 2
is_vowel= lambda ch : True  if(ch == "a" or ch == "e" or ch == "i" or ch == "o" or ch == "u") else False
def count_vowels(s):
    count = 0
    lenght = len(s)
    for i in range(lenght):
        ch = s[i].lower()
        if is_vowel(ch):
            count += 1
    return count
print(count_vowels("Hello World"))  # 3Z
````

**Problem 3**
Implement a simplified Run-Length Encoding (RLE):

Given a string s (letters only), compress it by replacing consecutive runs with:
<char><count>

Example:

"aaabbc" -> "a3b2c1"

"HHHaa" -> "H3a2"

Requirements:

Use a normal function flush_run(ch, count) that returns the encoded part for that run

Use a lambda append_piece that concatenates safely (just returns a + b)

Use a function rle_encode(s) returning the compressed string

````python
#Problem3
def flush_run(ch,count):
    return ch+str(count)
append_piece = lambda a,b:a+b
def rle_encode(s):
    if s == "":
        return s
    ch = s[0]
    count = 1
    result = ""
    lenght = len(s)
    for i in range(lenght):
        if i == 0:
            continue
        else:
            if ch == s[i]:
                count += 1
            else:
                temp = flush_run(ch,count)
                result = append_piece(result,flush_run(ch,count))
                ch = s[i]
                count = 1
    temp = flush_run(ch,count)
    result = append_piece(result,temp)
    return result
print(rle_encode("aaabbc"))   # a3b2c1
print(rle_encode("HHHaa"))    # H3a2
print(rle_encode(""))         # ""
````
