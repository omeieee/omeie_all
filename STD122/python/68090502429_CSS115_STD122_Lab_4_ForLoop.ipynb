## **For Loop**
The for loop is used to iterate a sequence range of the items of different data structures like list, tuple, string, etc.

**Syntax**


```
for value in sequence-range:
   loop body
```

1.   For keyword.
2.   The variable after the for keyword is the control variable of the loop, automatically counts the loop turns.
3. The in keyword.
4. The range () function.

````python
#range with one argument
for seq in range(10):
  print(seq)
````

````python
#range with two argument
for seq in range(5,10):
  print(seq)
````

````python
#range with three argument and argument is ascending
for seq in range(50,1000,-100):
  print(seq)
````

````python
#range with three argument and argument is descending
for seq in range(100,10,-1):
  print(seq)
````

````python
#negative range()
for i in range(-1, -11, -1):
  print(i, end=', ')
````

## **Nested For Loops**
The nested for loop is used to iterate different data structures and iterable data objects.

**Syntax**


```
# outer for loop
for element in sequence
   # inner for loop
   for element in sequence:
      body of inner for loop
   body of outer for loop
```



````python
n=int(input('enter number:'))
for i in range(1,n+1):
  for j in range(1,i+1):
    print(i,end=' ')
  print()
````

````python
n=int(input('enter number:'))
for i in range(1,n+1):
  for j in range(1,i+1):
    print(j,end=' ')
  print()
````

# **Using Else Statement with For Loop**
If the for loop contains any of the break statement, then the else statement will not be executed.

**Syntax**


```
for statement
# For loop code
else:
# Else block code
```




````python
for i in range(1, 5):
  print(i)
else: # Executed because no break in for
  print('No Break')
````

# **The Pass Statement in For Loop**
If we want to implement the code in the future and now, it should not be empty. If we empty for loop, then the interpreter throws an error. To prevent such type of error, pass will be used. The pass statement constructs the body of the for loop that does nothing.

Syntax


```
for val in sequence:
  pass
```



````python
for i in 'this is to test':
  if i=='e' or i=='o':
    pass
  else:
    print(i,end='')
````

# **Break Statement**

````python
for val in 'string':
  if val == 'i':
    break
  print(val)
````

````python
n = 1
while n < 5:
  n += 1
  if n == 3:
    break
  print(n)
````

````python
#break inner loop while
while True:
  print('In outer loop')
  i = 0
  while True:
    print('In inner loop')
    if i >= 5:
      break
    i += 1
  print('Got out of inner loop, still inside outer loop')
  break
````

# **Continue**

````python
# Program to show the use of continue statement inside loops
for val in 'string':
  if val == 'i':
    continue
  print(val)
````

````python
first = [1,2,3]
second = [4,5]
for i in first:
  for j in second:
    if i == j:
      continue
    print(i, '*', j, '=', i * j)
````

````python
i=1
while i<=2 :
  print(i,'outer')
  j=1
  while j<=2:
    print(j,'Inner')
    j+=1
    if(j==2):
      continue
  i+=1;
````

# **Problem 3**
