# **Decision-Making and Conditionals**

Conditional statements can check the conditions and accordingly change the behavior of the program.
Python programming language supports different decision-making statements:



*   `If` statement
*   `If-else` statement
* `Nested if` statement
* `elif` statement


# **`If` Statement**

The condition after the if statement is the Boolean expression. If the condition becomes true, then the specified block of statement runs, otherwise nothing happens.

**Syntax**


```
if condition:

  # Execute the indented block if true

  Statement-1

  Statement-2

  Statement-3

  Statement-4
```

Necessary elements for the if statement:

1.   The `if` keyword.
2.   One or more white spaces.
3. An expression whose value is intercepted as true or false. If the expression is interpreted as true, then the indented block will be executed. If the expression that is evaluated is false, the indented block will be omitted, and the next instruction after the indented block level will be executed.
4. A colon followed by a newline.
5. An indented set of statements.

````python
a,b=5,10
if(a<b):
  print("a is smaller than b")
if(b<a):
  print("b is smaller than a")
````

#`If-Else` Statement

If the condition evaluates to true, then the if block will be executed otherwise the else block will be executed.

**Syntax**

```
if condition:
  #Execute the indented block if true
  Statement-1
  Statement-2
else:
  #Execute the indented block if condition meets false
  Statement-3
  Statement-4
```

**Example:** Find the maximum of two numbers

````python
num1 = int(input("Enter the num1:"))
num2 = int(input("Enter the num2:"))
if num1 >= num2:
  print("The maximum of all values is",(num1))
else:
  print("The maximum of all values is",(num2))
````

# **Elif Statement**
**Syntax**


```
If condition 1:
  Statements-1
elif condition 2:`
  # Execute the indented block if the condition-1 becomes false
  Statements-2
elif condition 3:
  # Execute the indented block if all the above two conditions become false
elif condition-n:
  # Execute the indented block if all the above n-1 conditions become false
  Statements-n
else:
  # Execute the indented block if all the above conditions become false
  statement
```

In the elif statement, else is always the last branch, and the else block is an optional block.

**Example:** Calculating Grade

````python
m1=int(input("enter m1:"))
m2=int(input("enter m2:"))
m3=int(input("enter m3:"))
p=int((m1+m2+m3)/3)
if(p>90):
  print("Grade-A")
elif (p>80 and p<=90):
  print("Grade-B")
elif (p>60 and p<=80):
  print("Grade-C")
elif (p>60 and p<=45):
  print("Pass")
else:
  print("Fail")
````

# **While**
While loop repeats the execution if the condition evaluates to **true**.

If the condition is **false** for the first time, the while loop body is not executed even once.

**Syntax**


```
while condition:
  # Indent block is executed if the condition evaluates to true
  statement 1
  statement 2
  statement 3
```

**Example:** Find smallest of
three integers without comparison operators

````python
def smallest(x, y, z):
  c = 0
  while (x and y and z) :
    x = x-1
    y = y-1
    z = z-1
    c = c + 1
  return c
x = 12
y = 15
z = 5
print("Minimum of 3 numbers is", smallest(x, y, z))
````

**Break Statement**

Exists the loop immediately and unconditionally ends the loop; the control goes to the next instruction after the loop body.

````python
n = 1
while n < 5:
  if n == 3:
    break
  print(n)
  n += 1
````

**Continue**

````python
i = 1
while i <= 5:
  i = i+1
  if(i==3):
    continue
  print(i)
````

**Problem 4:**

Generate the Pascal's Triangle up to a given number of rows using nested while loops.

Pascal’s Triangle is a triangular array where:



*   The first and last values in each row are 1.
*   Each middle value is the sum of the two numbers directly above it.

For example, Pascal's Triangle with 5 rows looks like:


```
        1   
      1   1   
    1   2   1   
  1   3   3   1   
1   4   6   4   1   

```



````python
#Problem 4
rows = int(input("Enter number of rows: "))
row = 0

while row < rows:
    print("  " * (rows - row), end="")
    colum = 0
    value = 1  
    
    while colum <= row:
        print(value, end="   ")
        value = value * (row - colum) // (colum + 1)     
        colum += 1
        
    print()
    row += 1
````
