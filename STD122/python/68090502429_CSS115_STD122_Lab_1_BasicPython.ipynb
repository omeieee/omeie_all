# Print () Function

The `print()` prints a message onto the screen or to the standard output.

**Syntax:**

`print(object, sep=' ', end= '\n', file=sys.stdout, flush=False)`



`sep` – Specifies the separator between the outputted argument.

`end` – Specifies what to print at the end of the print statement

`flush` – Flushes the screen. Its default value is false

`file` – Where to display the message object. The default out is to the console, that is, std.out.

````python
print("this is to test",end=',')
print("this is the second line in python program")
print('test3')
````

````python
this=1
test=2
kuy = 3

print(this,test,kuy,sep=" hee ",end="... ")
print(f'this is {this} this is {test}')
````

````python
print("this is to test")
print(end="\n")
print("this is the second line in python program")
````

````python
print("this","is","to","test",sep=",")
print("this is the second line in python program")
````

````python
print("this is", "to test",sep="---",end="#")
print(end="\n")
print("this is an experiment")
````

````python
print("this","is","to,test",end="\n\n\n")
print("this is an experiment")
````

# Python Variables

**Naming Conventions to Variables**

*   The name of the variables must be composed of uppercase or lowercase alphabets (A … Z, a … z), digits (0 … 9) and the special symbol underscore (_).
*   The name of the variable must begin with a letter or the special symbol underscore (_) but not with a digit.
* The name of the variables must be a reserved keyword.
* Uppercase and lowercase letters are treated as different in Python.
* Python does not have any restrictions on the length of the variable name.
* Some valid variable names are test, sum, test_sum, test1, test1_avg, myVariable.
* Some invalid variable names are 5sum, test sum, False, None.

**Syntax**

variable_name=value
