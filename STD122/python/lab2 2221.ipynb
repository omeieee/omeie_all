# **Example**

Calculate area and perimeter of the circle with radius given by the user.

```python
import math
radius = float(input("Enter the radius of the circle: "))
area = math.pi * radius ** 2
perimeter = 2 * math.pi * radius
print("The area of the circle is", area)
print("The perimeter of the circle is", perimeter)
```

**Problem 1**

Write a Python program to calculate the gross monthly salary of an employee. Take salary per hour, number of hour from user and add special 3% topup as a bonus.

```python
#Problem 1
salary=int(input("please enter your salary : "))
hour=int(input("please enter your hour : "))
salary_totle = salary * hour 
salary_totle = salary_totle + (salary_totle * 3/100)
print(f"your totle salary is : {salary_totle}")
```

**Problem 2**

Find the roots of a quadratic equation by given `a`, `b`, `c` by user
ex.
`a = 1, b = -7, c = 10`
Roots are 2 and 5.



```python
#Problem 2
import math
a = int(input("Enter a : "))
b = int(input("Enter b : "))
c = int(input("Enter c : "))

d = (b**2) - (4*a*c)

root1 = (-b + math.sqrt(d)) / (2*a)
root2 = (-b - math.sqrt(d)) / (2*a)

print(f"Roots are {root1} and {root2}")
```

**Problem 3**
A vending machine can return coins in denominations of 1, 2, 5, and 10 baht. Write a program that calculates the number of each coin denomination the machine will dispense for a given amount (in baht). Always start with the highest denomination (10 baht).

**Problem 4**
Write a Python program that calculates and prints the average of five numbers entered by the user.

```python
#Problem 4
print("Please input 5 number")
number1 = float(input("Enter your First number : "))
number2 = float(input("Enter your Second number : "))
number3 = float(input("Enter your Third number : "))
number4 = float(input("Enter your Fourth number : "))
number5 = float(input("Enter your Fifth number : "))

average = (number1 + number2 + number3 + number4 + number5) / 5

print(f"Your average of five numbers is : {average}")
```
