```python
#Problem 4
def calc_numBox(cap1,sum1):
    total = sum1 // cap1
    return total

cap_w = int(input("Enter your wight cap: "))
n = int(input("Enter your num of cap: "))
sum = 0
for i in range(n):
    w = int(input(f"Enter your wight {i+1}"))
    sum += w
    i += 1
print("Total cap is:",calc_numBox(cap_w,sum))
```

```python
#Problem 5
def remaining(principal,rate,months,pay):
    principal = principal * (1 + rate/100)
    principal = principal - pay
    return principal


principal = int(input("Enter your principal"))
rate = int(input("Enter your rate"))
months = int(input("Enter your months"))
pay = int(input("Enter your pay"))
print("Toalt is:",remaining(principal,rate,months,pay))
```
