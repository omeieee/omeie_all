```python
#Problem 1

cars = ["Isuzu", "Nissan", "Mitsubishi", "Toyota", "Ford", "Bmw", "Audi", "Hyundai", "Kia", "Chevrolet"]
car_upper = list(map(lambda x: x.upper(), filter(lambda x: len(x) >= 6, cars)))
print(car_upper)
```

```python
#Problem 2

import statistics
number = input("Enter number separated by spaces: ")
num_split = number.split()
mean = statistics.mean(map(float, num_split))
median = statistics.median(map(float, num_split))
mode = statistics.mode(map(float, num_split))

print("The numbers you entered are:", number)
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
#Problem 3

number = input("Enter number separated by spaces: ")
num_split = number.split()
print("Enter number separated by spaces: ",number)
powers = list(map(lambda x: int(x)**2, num_split))
total_sum = sum(powers)
print("Squared number: ",powers)
print("Sum of squares: ",total_sum)
```
